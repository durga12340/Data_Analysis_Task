```python
  import pandas as pd

  # Load the CSV file
  df = pd.read_csv(r'SalesData/sales_data.csv')
  
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df.fillna(0, inplace=True)  # Example: fill NaN with 0
```

```python
grouped_data = df.groupby(' "Weight(Pounds)"').sum()
print(grouped_data)
  
```

```python
  import matplotlib.pyplot as plt

  grouped_data.plot(kind='bar')
  plt.title('Sales Data Analysis')
  plt.xlabel('Category')
  plt.ylabel('Total Sales')
  plt.show()
  
```
